```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the data
#replicate 1 link: https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSM5687481
#replicate 2 link: https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSM5687482
bulk = pd.read_csv('10x_combined_pseudobulk_both_replicates.txt', sep='\t', header=None, names=['gene', 'count'])
bulk.set_index('gene', inplace=True)

pseudo = pd.read_csv('scK562_NegCTRL_2X_Cells_ChronoSeq_Pseudobulk.txt', sep='\t', header=None, names=['gene', 'count'])
pseudo.set_index('gene', inplace=True)

# Merge on common genes
merged = bulk.join(pseudo, how='inner', lsuffix='_bulk', rsuffix='_pseudo')

# CPM normalization
bulk_cpm = (merged['count_bulk'] * 1e6) / merged['count_bulk'].sum()
pseudo_cpm = (merged['count_pseudo'] * 1e6) / merged['count_pseudo'].sum()

# Log2 CPM transformation
log2_bulk = np.log2(bulk_cpm + 1)
log2_pseudo = np.log2(pseudo_cpm + 1)

# Currently no genes filtered
keep_mask = (log2_bulk > 0) & (log2_pseudo > 0)

x = log2_bulk[keep_mask].values
y = log2_pseudo[keep_mask].values

# Calculate correlation
r, p = pearsonr(x, y)

# Subsample for plotting (10,000 points)
n = min(10000, len(x))
np.random.seed(42)  # For reproducibility
idx = np.random.choice(len(x), n, replace=False)
x_plot = x[idx]
y_plot = y[idx]

# Create high-resolution figure
fig, ax = plt.subplots(figsize=(10, 10), dpi=300)  # High resolution (300 DPI)

# Scatter plot with semi-transparent points
ax.scatter(x_plot, y_plot, alpha=0.6, color='blue', s=10)  # Smaller point size for density

# Diagonal reference line (red)
min_val = min(ax.get_xlim()[0], ax.get_ylim()[0])
max_val = max(ax.get_xlim()[1], ax.get_ylim()[1])
ax.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', linewidth=1)

# Set same scale for both axes
ax.set_xlim(0, 16)
ax.set_ylim(0, 16)

# Add grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Annotation for correlation
ax.text(1, 14, f'r = {r:.2f}', fontsize=14, fontweight='bold')

# Labels and title
ax.set_xlabel('10X Single-Cell Pseudobulk both replicates combined(log2 CPM)', fontsize=12)
ax.set_ylabel('ChronoSeq Single-Cell Pseudobulk all splits (log2 CPM)', fontsize=12)
ax.set_title('ChronoSeq K562 Single-Cell Negative Control Pseudobulk vs 10X Genomics Single-Cell Pseudobulk', fontsize=12)

# Export as SVG
plt.savefig('ChronoSeq_Single_Cell_all_splits_Pseudobulk_vs_10X_Single_Cell_both_replicates_pseudobulk.svg', format='svg', dpi=300)  # High-resolution SVG export

# Show plot
plt.tight_layout()
plt.show()
```
